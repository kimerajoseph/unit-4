```python
# 7. Define a function that differentiates between continuous and discrete variables.
# COLUMN CN BE NUMERICAL AND DISCRETE OR CTEGORICAL AND CONTINOUS
# IF A COLUMN HAS MORE THAT 10 UNIQUES VALUES, THEN ITS CONTINOUS AND THE REVERSE IS TRUE

continuous, discrete = [],[]
def diferentiate():
    for col in customer_df.columns:
        if len(customer_df[col].unique()) < 10:
            #print("DISCRETE")
            discrete.append(col)
        
        else:
            #print("CONTINOUS")
            continuous.append(col)
    print("CONTINOUS COLUMNS: ",continuous)
    print("DISCRETE COLUMNS: ",discrete)

# CALL THE FUNCTION
diferentiate()    
```

```python
# 8. For the categorical data, check if there is some kind of text in a variable so we would need to clean it.
# All discrete categorical data values are strings. I will search for strings only in the numerical columns

customer_discrete = customer_df[discrete]
display(customer_discrete.head())

numerical_categoric = customer_discrete.select_dtypes(np.number)
numerical_categoric.head()
```
